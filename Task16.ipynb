```python

# cчитаем выбросы значений ректальной температуре:
q1 = df8.rectal_temperature.quantile(0.25)
q3 = df8.rectal_temperature.quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - (1.5 * iqr) 
upper_bound = q3 + (1.5 * iqr)
remove_outliers_temp = df8[df8.rectal_temperature.between(lower_bound, upper_bound, inclusive=True)].sort_values('rectal_temperature')
remove_outliers_temp

```

```python
# cчитаем выбросы значений пульса:
q1 = df8.pulse.quantile(0.25)
q3 = df8.pulse.quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - (1.5 * iqr) 
upper_bound = q3 + (1.5 * iqr)
remove_outliers_pulse = df8[df8.pulse.between(lower_bound, upper_bound, inclusive=True)].sort_values('pulse')
remove_outliers_pulse
```
